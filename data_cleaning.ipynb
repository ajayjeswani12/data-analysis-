**DATA CLEANING IN PYTHON**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from ast import literal_eval
import json
%matplotlib inline
```

```python
df = pd.read_csv("movie.csv", low_memory=False)
```

```python
df.shape
df
df.info()
```

```python
drop_df = ["title_year.1"]
df = df.drop(drop_df, axis=1) # drops the selected columns
df = df.drop_duplicates(keep='first') # removes the duplicates from existing dataframe
df.dropna(how="all",inplace=True) # if each column is NaN or null in a row, drops this row
```

```python
df.shape
df.info()
```

```python
df["duration"] = df["duration"].replace(0, np.nan)
df["duration"].fillna(df["duration"].mean(), inplace=True) 
```

```python
df["num_voted_users"] = df["num_voted_users"].replace(0, np.nan)
df["num_voted_users"].fillna(df["num_voted_users"].mean(), inplace=True) 
```

```python
face_med =   df['facenumber_in_poster'].median()
face_med
df["facenumber_in_poster"].fillna(face_med, inplace = True)
```

```python
total_cast_med =   df['Cast_Total_facebook_likes'].median()
total_cast_med
df["Cast_Total_facebook_likes"].fillna(total_cast_med, inplace = True)
```

```python
df["ACTOR_2_facebook_likes"] = df["ACTOR_2_facebook_likes"].replace(0, np.nan)
df["ACTOR_2_facebook_likes"].fillna(df["ACTOR_2_facebook_likes"].mean(), inplace=True) 
```

```python
df['DIRECTOR_facebook_likes'] = df['DIRECTOR_facebook_likes'].str.replace('"475"','475')
print (df)
```

```python
df["DIRECTOR_facebook_likes"] =pd.to_numeric(df['DIRECTOR_facebook_likes'], errors='coerce', downcast="float") 
director_med =   df['DIRECTOR_facebook_likes'].median()
director_med
df["DIRECTOR_facebook_likes"].fillna(director_med, inplace = True)
```
